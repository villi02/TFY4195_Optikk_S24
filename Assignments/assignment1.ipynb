```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1)

```python
# Define constants
n_water = 1.334
n_air = 1.000


# Define functions
# Calculate the angle of refraction from water to air
def refraction_angle_external_reflection(n1, n2, theta1):
    return np.arcsin(n1/n2 * np.sin(theta1))

# Calculate the angle of refraction from air to water
def refraction_angle_internal(n1, n2, theta2):
    return np.arcsin(n2/n1 * np.sin(theta2))

# Calculate TE reflection coefficient
def reflection_coefficient_TE(n1, n2, theta1, theta2):
    return (n1*np.cos(theta1) - n2*np.cos(theta2)) / (n1*np.cos(theta1) + n2*np.cos(theta2))

# Calculate TM reflection coefficient
def reflection_coefficient_TM(n1, n2, theta1, theta2):
    return (n2*np.cos(theta1) - n1*np.cos(theta2)) / (n2*np.cos(theta1) + n1*np.cos(theta2))

# Calculate TE transmission coefficient
def transmission_coefficient_TE(n1, n2, theta1, theta2):
    return 2*n1*np.cos(theta1) / (n1*np.cos(theta1) + n2*np.cos(theta2))

# Calculate TM transmission coefficient
def transmission_coefficient_TM(n1, n2, theta1, theta2):
    return 2*n1*np.cos(theta1) / (n2*np.cos(theta1) + n1*np.cos(theta2))

# Calculate the reflected light intensity
def reflected_intensity(r_TE, r_TM):
    return (r_TE**2 + r_TM**2) / 2

# Calculate the transmitted light intensity
def transmitted_intensity(t_TE, t_TM):
    return (t_TE**2 + t_TM**2) / 2

# Calculate the Brewster angle
theta_B = np.arctan(n_air/n_water)

# Calculate the critical angle
theta_C = np.arcsin(n_air/n_water)

```

## 1a) - external reflection - Plotting 

```python
# Plot the reflection and transmission coefficients for air to water

# Define the angles
theta1 = np.linspace(0, np.pi/2, 1000)
theta2_external = refraction_angle_external_reflection(n_air, n_water, theta1)

# Calculate the reflection and transmission coefficients from air to water
r_TE = reflection_coefficient_TE(n_air, n_water, theta1, theta2_external)
t_TE = transmission_coefficient_TE(n_air, n_water, theta1, theta2_external)
r_TM = reflection_coefficient_TM(n_air, n_water, theta1, theta2_external)
t_TM = transmission_coefficient_TM(n_air, n_water, theta1, theta2_external)

# Plot the reflection and transmission coefficients
plt.figure()
plt.plot(theta1, r_TE, label='Reflection TE')
plt.plot(theta1, r_TM, label='Reflection TM')
plt.plot(theta1, t_TE, label='Transmission TE')
plt.plot(theta1, t_TM, label='Transmission TM')

# Add vertical lines for the Brewster angle and the critical angle
plt.axvline(x=theta_B, color='grey', linestyle='--', label='Brewster angle')
plt.axvline(x=theta_C, color='black', linestyle='-.', label='Critical angle')

# Add labels and legend
plt.xlabel('Incident angle (rad)')
plt.ylabel('Reflection coefficient')
plt.legend()
plt.show()
```

## 1b) - internal reflection - Plotting 

```python
# Plot the reflection and transmission coefficients for water to air

# Define the angles
theta1 = np.linspace(0, np.pi/2, 1000)
theta2 = refraction_angle_external_reflection(n_water, n_air, theta1)

# Calculate the reflection and transmission coefficients for water to air
r_TE = reflection_coefficient_TE(n_water, n_air, theta1, theta2)
t_TE = transmission_coefficient_TE(n_water, n_air, theta1, theta2)
r_TM = reflection_coefficient_TM(n_water, n_air, theta1, theta2)
t_TM = transmission_coefficient_TM(n_water, n_air, theta1, theta2)

# Plot the reflection and transmission coefficients
plt.figure()
plt.plot(theta1, r_TE, label='Reflected TE')
plt.plot(theta1, r_TM, label='Reflected TM')
plt.plot(theta1, t_TE, label='Transmitted TE')
plt.plot(theta1, t_TM, label='Transmitted TM')

# Add vertical lines for the Brewster angle and the critical angle
plt.axvline(x=theta_B, color='grey', linestyle='--', label='Brewster angle')
plt.axvline(x=theta_C, color='black', linestyle='-.', label='Critical angle')

# Add labels and legend
plt.xlabel('Incident angle (rad)')
plt.ylabel('Reflection coefficient')
plt.legend()
plt.show()
```

## Light intensity for external and internal reflection for TE and TM polarizations

```python
# Plot the reflected and transmitted light intensity for external and internal reflection

# Define the angles
theta1 = np.linspace(0, np.pi/2, 1000)
theta2_external = refraction_angle_external_reflection(n_air, n_water, theta1)
theta2_internal = refraction_angle_internal(n_air, n_water, theta1)

# Calculate the reflection and transmission coefficients for external and internal reflection
r_TE_external = reflection_coefficient_TE(n_air, n_water, theta1, theta2_external)
t_TE_external = transmission_coefficient_TE(n_air, n_water, theta1, theta2_external)
r_TM_external = reflection_coefficient_TM(n_air, n_water, theta1, theta2_external)
t_TM_external = transmission_coefficient_TM(n_air, n_water, theta1, theta2_external)

r_TE_internal = reflection_coefficient_TE(n_air, n_water, theta1, theta2_internal)
t_TE_internal = transmission_coefficient_TE(n_air, n_water, theta1, theta2_internal)
r_TM_internal = reflection_coefficient_TM(n_air, n_water, theta1, theta2_internal)
t_TM_internal = transmission_coefficient_TM(n_air, n_water, theta1, theta2_internal)

# Calculate the reflected and transmitted light intensity
reflected_intensity_external = reflected_intensity(r_TE_external, r_TM_external)
transmitted_intensity_external = transmitted_intensity(t_TE_external, t_TM_external)
reflected_intensity_internal = reflected_intensity(r_TE_internal, r_TM_internal)
transmitted_intensity_internal = transmitted_intensity(t_TE_internal, t_TM_internal)

# Plot the reflected and transmitted light intensity
plt.figure(figsize=(10, 5))
plt.plot(theta1, reflected_intensity_external, label='Reflected intensity external')
plt.plot(theta1, transmitted_intensity_external, label='Transmitted intensity external')
plt.plot(theta1, reflected_intensity_internal, label='Reflected intensity internal')
plt.plot(theta1, transmitted_intensity_internal, label='Transmitted intensity internal')

# Add vertical lines for the Brewster angle and the critical angle
plt.axvline(x=theta_B, color='grey', linestyle='--', label='Brewster angle')
plt.axvline(x=theta_C, color='black', linestyle='-.', label='Critical angle')

# Add labels and legend
plt.xlabel('Incident angle (rad)')
plt.ylabel('Intensity')
plt.legend()
```
