```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1)

```python
# Define constants
n_water = 1.334
n_air = 1.000


# Define functions
# Calculate the angle of refraction from water to air
def refraction_angle_external_reflection(n1, n2, theta1):
    return np.arcsin(n1/n2 * np.sin(theta1))

# Calculate the angle of refraction from air to water
def refraction_angle_internal(n1, n2, theta2):
    return np.arcsin(n2/n1 * np.sin(theta2))

# Calculate TE reflection coefficient
def reflection_coefficient_TE(n1, n2, theta1, theta2):
    return (n1*np.cos(theta1) - n2*np.cos(theta2)) / (n1*np.cos(theta1) + n2*np.cos(theta2))

# Calculate TM reflection coefficient
def reflection_coefficient_TM(n1, n2, theta1, theta2):
    return (n2*np.cos(theta1) - n1*np.cos(theta2)) / (n2*np.cos(theta1) + n1*np.cos(theta2))

# Calculate TE transmission coefficient
def transmission_coefficient_TE(n1, n2, theta1, theta2):
    return 2*n1*np.cos(theta1) / (n1*np.cos(theta1) + n2*np.cos(theta2))

# Calculate TM transmission coefficient
def transmission_coefficient_TM(n1, n2, theta1, theta2):
    return 2*n1*np.cos(theta1) / (n2*np.cos(theta1) + n1*np.cos(theta2))

# Calculate the reflected light intensity
def reflected_intensity(r_TE, r_TM):
    return (r_TE**2 + r_TM**2) / 2

# Calculate the transmitted light intensity
def transmitted_intensity(t_TE, t_TM):
    return (t_TE**2 + t_TM**2) / 2

# Calculate the Brewster angle
def brewster_angle_internal(n1, n2):
    return np.arctan(n2/n1)


# Calculate the critical angle
def critical_angle_internal(n1, n2):
    return np.arcsin(n2/n1)

# Calculate the critical angle
def critical_angle_external(n1, n2):
    return np.arcsin(n2/n1)
```

## 1a) - external reflection - Plotting 

```python
# Plot the reflection and transmission coefficients for air to water

# Define the angles
theta1 = np.linspace(0, np.pi/2, 1000)
theta2_external = refraction_angle_external_reflection(n_air, n_water, theta1)

# Calculate brewster angle
theta_B = brewster_angle_internal(n_air, n_water)
theta_C = critical_angle_internal(n_air, n_water)

# Calculate the reflection and transmission coefficients from air to water
r_TE = reflection_coefficient_TE(n_air, n_water, theta1, theta2_external)
t_TE = transmission_coefficient_TE(n_air, n_water, theta1, theta2_external)
r_TM = reflection_coefficient_TM(n_air, n_water, theta1, theta2_external)
t_TM = transmission_coefficient_TM(n_air, n_water, theta1, theta2_external)

# Plot the reflection and transmission coefficients

# Reflection
plt.figure()
plt.plot(theta1, r_TE, label='Reflection TE')
plt.plot(theta1, r_TM, label='Reflection TM')


# Add vertical lines for the Brewster angle and the critical angle
plt.axvline(x=theta_B, color='grey', linestyle='--', label='Brewster angle')
plt.axvline(x=theta_C, color='black', linestyle='-.', label='Critical angle')

# Add labels and legend
plt.xlabel('Incident angle (rad)')
plt.ylabel('Reflection coefficient')
plt.legend()
plt.show()

## Transmission
plt.figure()
plt.plot(theta1, t_TE, label='Transmission TE')
plt.plot(theta1, t_TM, label='Transmission TM')

# Add vertical lines for the Brewster angle and the critical angle
plt.axvline(x=theta_B, color='grey', linestyle='--', label='Brewster angle')
plt.axvline(x=theta_C, color='black', linestyle='-.', label='Critical angle')

# Add labels and legend
plt.xlabel('Incident angle (rad)')
plt.ylabel('Reflection coefficient')
plt.legend()
plt.show()
```

## 1b) - internal reflection - Plotting

```python

# Constants
n_air = 1.000
n_water = 1.334

# Your functions are defined here...

# Define incident angle array
theta1 = np.linspace(0, np.pi/2, 500)

# Helper function to calculate theta2 for each theta1
def calculate_theta2(n1, n2, theta1_array, is_external):
    theta2_array = np.zeros_like(theta1_array)
    for i, theta1 in enumerate(theta1_array):
        if is_external:
            theta2_array[i] = refraction_angle_external_reflection(n1, n2, theta1)
        else:
            theta2_array[i] = refraction_angle_internal(n1, n2, theta1)
    return theta2_array

# Calculate theta2 for external and internal reflections
theta2_ext = calculate_theta2(n_air, n_water, theta1, True)
theta2_int = calculate_theta2(n_water, n_air, theta1, False)

# Calculate reflection and transmission coefficients
r_TE_ext = np.array([reflection_coefficient_TE(n_air, n_water, theta, theta2) for theta, theta2 in zip(theta1, theta2_ext)])
r_TM_ext = np.array([reflection_coefficient_TM(n_air, n_water, theta, theta2) for theta, theta2 in zip(theta1, theta2_ext)])
t_TE_ext = np.array([transmission_coefficient_TE(n_air, n_water, theta, theta2) for theta, theta2 in zip(theta1, theta2_ext)])
t_TM_ext = np.array([transmission_coefficient_TM(n_air, n_water, theta, theta2) for theta, theta2 in zip(theta1, theta2_ext)])

r_TE_int = np.array([reflection_coefficient_TE(n_water, n_air, theta, theta2) for theta, theta2 in zip(theta1, theta2_int)])
r_TM_int = np.array([reflection_coefficient_TM(n_water, n_air, theta, theta2) for theta, theta2 in zip(theta1, theta2_int)])
t_TE_int = np.array([transmission_coefficient_TE(n_water, n_air, theta, theta2) for theta, theta2 in zip(theta1, theta2_int)])
t_TM_int = np.array([transmission_coefficient_TM(n_water, n_air, theta, theta2) for theta, theta2 in zip(theta1, theta2_int)])

# Calculate intensities
R_ext = reflected_intensity(r_TE_ext, r_TM_ext)
T_ext = transmitted_intensity(t_TE_ext, t_TM_ext)
R_int = reflected_intensity(r_TE_int, r_TM_int)
T_int = transmitted_intensity(t_TE_int, t_TM_int)

# Brewster and critical angles
brewster_angle = brewster_angle_internal(n_air, n_water)
critical_angle = critical_angle_internal(n_water, n_air)

# Plotting
plt.figure(figsize=(12, 8))

# External reflection and transmission
plt.subplot(2, 2, 1)
plt.plot(theta1, R_ext, label='Reflected Intensity')
plt.plot(theta1, T_ext, label='Transmitted Intensity')
plt.axvline(brewster_angle, color='grey', linestyle='--', label=f'Brewster Angle = {brewster_angle:.2f} rad')
plt.title('External Reflection and Transmission')
plt.xlabel('Incident Angle (rad)')
plt.ylabel('Intensity')
plt.legend()

# Internal reflection and transmission
plt.subplot(2, 2, 2)
plt.plot(theta1, R_int, label='Reflected Intensity')
plt.plot(theta1, T_int, label='Transmitted Intensity')
plt.axvline(critical_angle, color='red', linestyle='-.', label=f'Critical Angle = {critical_angle:.2f} rad')
plt.title('Internal Reflection and Transmission')
plt.xlabel('Incident Angle (rad)')
plt.legend()

plt.tight_layout()
plt.show()

```
