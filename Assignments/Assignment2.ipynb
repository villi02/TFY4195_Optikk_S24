```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## 1)

### Single slit

```python
# Define constants
lambda_0 = 532e-9 # Wavelength of light in meters
d_0 = 2e-6 # Distance between slits in meters
l_0 = 4e-6 # Distance from slits to screen in meters
x = np.linspace(-1, 1, 100) # Define x values
```

```python
# Define the intensity
def normalized_intensity(x=x, lambda_=lambda_0, d=d_0, y=1):
    beta = np.pi * d * x / (lambda_ * y)
    #underneath = (np.pi*d*x/(lambda_*y))
    intensity = (np.sinc(beta/np.pi)) ** 2
    return intensity
```

#### As function of x

```python
# Plot the intensity
values = normalized_intensity(x, lambda_0, d_0, 1)
plt.plot(x, values)
plt.xlabel('x')
plt.ylabel('Intensity')
plt.title('Intensity of single slit')
plt.show()
```

#### As function of D

```python
# Plot the intensity

"""
d = np.linspace(1e-6, 10e-6, 1000)

values = normalized_intensity(1, lambda_0, d, 1)
plt.plot(d, values)
plt.xlabel('d')
plt.ylabel('Intensity')
plt.title('Intensity of single slit, varying d')
plt.show()
"""

# make an array of 6 different values of d
d = np.linspace(1e-6, 10e-6, 6)

# plot the intensity for each value of d
for i in d:
    values = normalized_intensity(x, lambda_0, i, 1)
    plt.plot(x, values, label=f'd={i:.2e}')
plt.xlabel('x')
plt.ylabel('Intensity')
plt.title('Intensity of single slit, varying d')
plt.legend()
```

#### As a function of lambda

```python
# Plot the intensity

lambda_ = np.linspace(400e-9, 700e-9, 6)

# plot the intensity for each value of lambda
for i in lambda_:
    values = normalized_intensity(x, i, d_0, 1)
    plt.plot(x, values, label=f'lambda={i:.2e}')
plt.xlabel('lambda')
plt.ylabel('Intensity')
plt.title('Intensity of single slit')
plt.show()
```

### Double slit

```python
# Define the intensity of the double slit
def normalized_intensity_double(x=1, lambda_=lambda_0, d=d_0, l=l_0, y=1):
    beta = np.pi * d * x / (lambda_ * y)
    beta_2 = np.pi * l * x / (lambda_ * y)
    intensity = (np.sinc(beta / np.pi)) ** 2 * np.cos(beta_2) ** 2
    return intensity
```

#### As a function of x

```python
# Plot the intensity
values = normalized_intensity_double(x=x)
plt.plot(x, values)
plt.xlabel('x')
plt.ylabel('Intensity')
plt.title('Intensity of double slit')
plt.show()
```

#### As a function of l

```python
l = np.linspace(1e-6, 10e-6, 6)


# plot the intensity for each value of l
for i in l:
    values = normalized_intensity_double(x, lambda_0, d_0, i, 1)
    plt.plot(x, values, label=f'l={i:.2e}')
plt.xlabel('l')
plt.ylabel('Intensity')
plt.title('Intensity of double slit')
plt.show()
```

### Diffraction grating using Fraunhofer approximation

```python
# define intensity of a diffraction grating using Fraunhofer diffraction
def normalized_intensity_grating(x=1, lambda_=lambda_0, d=d_0, N=1, y=1):
    beta = np.pi * d * x / (lambda_ * y)
    beta_2 = np.pi * N * x / (lambda_ * y)
    intensity = (np.sinc(beta / np.pi)) ** 2 * (np.sin(beta_2) / beta_2) ** 2
    return intensity
```

#### As a function of N

```python
# Plot the intensity
N = np.linspace(1, 5, 5)

# plot the intensity for each value of N
for i in N:
    values = normalized_intensity_grating(x, lambda_0, d_0, N=i)
    plt.plot(x, values, label=f'N={i:.2e}')
plt.xlabel('x')
plt.ylabel('Intensity')
plt.title('Intensity of double slit')
plt.show()
```
