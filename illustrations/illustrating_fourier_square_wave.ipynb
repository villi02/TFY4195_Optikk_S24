```python
import numpy as np
import matplotlib.pyplot as plt

f = np.linspace(-10, 10, 1000)
t0 = 1  # example duration
A = 1  # example amplitude
sinc_function = (A * t0 * np.sinc(t0 * f))**2

plt.figure(figsize=(10, 5))
plt.plot(f, sinc_function)
plt.title('Power Spectrum of a Square Pulse')
plt.xlabel('Frequency')
plt.ylabel('Power')
plt.grid(True)
plt.show()
```
